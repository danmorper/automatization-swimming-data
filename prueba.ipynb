```python
def column_hyphens(df):
    """
    df: dataframe
    Output: column with more than 50% of hyphens
    We use this function in points function, in order to find the column we'll called score
    """
    for column in df.columns:
        hyphen_count = sum(df[column] == "-")
        if hyphen_count >= len(df) / 4:
            return column
    
    return hyphen_count

print(tabu27.columns)
```
